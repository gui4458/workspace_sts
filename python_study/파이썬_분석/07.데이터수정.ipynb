```python
import pandas as pd
df = pd.read_excel('student.xlsx')
df
```

```python
#데이터 교체(대체)
# 데이터 저장을 위해서는 inplace 속성 추가
df['학급'].replace({'자바반':'수학반','캐드반':'영어반'})

my_data = {
    '자바반' : '수학반'
    ,'캐드반' : '과학반'
}
df['학급'].replace(my_data)
```

```python
#이름 컬럼값 변경
df['이름'] = 'python'
df
```

```python
# 특정 조건이 만족하는 행과 열을 찾아서 데이터 변경
df.loc[df['학급'] == '캐드반','학급'] = '과학반'
df
```

```python
df['총점'] = df['국어점수']+df['영어점수']
df
```
