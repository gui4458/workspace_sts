```python
import bs4
```

```python
# response.text을 출력하면 웹 사이트의 내용을 '문자열로' 출력
print(response.text)
```

```python
# 문자열 데이터에서 우리가 필요한 데이터로 쉽게 접근하기 위해서
# BeautifulSoup 형태로 데이터 변환
# 아래 코드 실행 시 오류 발생 하면 'lxml' 설치 후 재실행
bs_data = bs4.BeautifulSoup(response.text, 'lxml')
print(bs_data)
```

```python
# find('태그명') -> 해당 태그 출력
# find('태그명').text -> 해당 태그 사이 내용만 찾음

# findAll('태그명') -> 해당 태그 전부 찾음
aa = bs_data.find('resultcode')
print(aa)
```

```python
# 모든 item 태그들의 정보만 가져옴
items = bs_data.findAll('item')
print(items)
```

```python
dataTime_list = []
for item in items : 
    print(item.find('datatime').text)
    dataTime_list.append(item.find('datatime').text)
# API로 가져온 데이터를 모두 저장할 딕셔너리
dic_data = {}
dic_data['dataTime'] = dataTime_list
dic_data
```
