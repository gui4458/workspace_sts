+ maria db 연결을 위해서 라이브러리를 설치

+ 설치한 라이브러리를 import 하여 사용 -> import pymysql

```python
import pymysql
```

```python
# 1. 파이썬에 디비 연결
conn = pymysql.connect(host='localhost',user='study_user',passwd='mariadb',db='study_db',charset='utf8')
```

```python
# 쿼리 조회하기

# 쿼리의 결과 및 실행되고 있는 위치 정보를 포함하는 객체
cursor = conn.cursor(pymysql.cursors.DictCursor)
# 실행할 sql
sql = 'SELECT * FROM EMP'

# 실제로 쿼리를 실행하는 함수
cursor.execute(sql)

# 조회된 모든 데이터를 조회
result1 = cursor.fetchall()
print(result1)
```

```python
for i in result1 :
    print(i)
```

```python
conn.close()
```
